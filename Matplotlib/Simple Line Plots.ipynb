This notebook deals with how to implement simple line plots in two different matplotlib styles, i.e. MATLAB and Object-Oriented

- basic line plotting
- line color and style
- axis limits
- title and labels: single and multiple legends

### Basic Line Plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1) # the same as fig.add_subplot(111)

x = np.linspace(0, 10, 1000)
ax1.plot(x, np.sin(x))
ax1.plot(x, np.cos(x))
```

### Adjusting the Plot: Line Colors and Styles

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1) # the same as fig.add_subplot(111)

ax1.plot(x, x + 0, linestyle='solid')
ax1.plot(x, x + 1, linestyle='dashed')
ax1.plot(x, x + 2, linestyle='dashdot')
ax1.plot(x, x + 3, linestyle='dotted')

ax1.plot(x, x + 4, linestyle='-') # solid
ax1.plot(x, x + 5, linestyle='--') # dashed
ax1.plot(x, x + 6, linestyle='-.') # dashdot
ax1.plot(x, x + 7, linestyle=':') # dotted
```

### Nonkeyword argument to plt.plot()
- RGB (Red/Green/Blue)
- CMYK (Cyan/Magenta/Yellow/blacK)

```python
plt.plot(x, x + 0, '-g') #solid green
plt.plot(x, x + 1, '--c') #dashed cyan
plt.plot(x, x + 2, '-.k') #dashdot black
plt.plot(x, x + 3, ':r') #dotted red
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1) # the same as fig.add_subplot(111)

ax1.plot(x, x + 0, '-g')
ax1.plot(x, x + 1, '--c')
ax1.plot(x, x + 2, '-.k')
ax1.plot(x, x + 3, ':r')
```

### Adjusting the Plot:Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# reverse axis
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1) # the same as fig.add_subplot(111)

ax1.plot(x, np.sin(x))
ax1.set_xlim(-1, 11)
ax1.set_ylim(-1.5, 1.5)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1) # the same as fig.add_subplot(111)

ax1.plot(x, np.sin(x))
ax1.axis([-1, 11,-1.5, 1.5])
#ax1.set(xlim=(-1, 11), ylim=(-1.5, 1.5))
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1) # the same as fig.add_subplot(111)

ax1.plot(x, np.sin(x))
ax1.axis('tight')
```

### Labeling Plots

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# legend for multiple lines (data)
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

ax1.plot(x, np.sin(x))
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.set_title('A Sine Curve')
```

### There are 3 different ways to set labels for multiple lines

```python
# 1. inside plot()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

ax1.plot(x, np.sin(x), '-g', label='sin(X)')
ax1.plot(x, np.cos(x), ':b', label='cos(X)')
ax1.axis('equal')
ax1.legend()
#ax1.legend('sin(x), cos(x)')
```

```python
# 2. using line ojbects
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

line1, = ax1.plot(x, np.sin(x), '-g')
line2, = ax1.plot(x, np.cos(x), ':b')
ax1.axis('equal')

line1.set_label('sin(x)')
line2.set_label('cos(x)')
ax1.legend()
```

```python
# 3. using legend()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

line1, = ax1.plot(x, np.sin(x), '-g')
line2, = ax1.plot(x, np.cos(x), ':b')
ax1.axis('equal')

#legend(handles, labels)
ax1.legend((line1, line2), ('sin(X)', 'cos(X)'))
```
