# Simple Scatter Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

### Scatter Plots with plt.plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

ax1.plot(x, y, 'o', color='black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
    label="marker='{0}'".format(marker))

plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

### Scatter Plots with plt.scatter

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

# each individual point can be individually controlled or mapped to data
plt.scatter(x, y, marker='o')
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.scatter(x, y, marker='o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar() # show color scale
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100) #randn : standard normal
colors = rng.rand(100) #rand : random samples from a uniform dist [0,1)
sizes = 1000 * rng.rand(100)

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

# how to generator colorbar data ?
cax = ax1.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
fig.colorbar(cax, ax=ax1)
```

### Color Bar Placement

#### To find colorbar placement, 
#### https://matplotlib.org/3.1.0/gallery/subplots_axes_and_figures/colorbar_placement.html


```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2)
cm = ['RdBu_r', 'viridis']
for col in range(2):
    for row in range(2):
        ax = axs[row, col]
        pcm = ax.pcolormesh(np.random.random((20, 20)) * (col + 1),
                            cmap=cm[col])
        fig.colorbar(pcm, ax=ax)
```

```python
fig, axs = plt.subplots(2, 2)
cm = ['RdBu_r', 'viridis']
for col in range(2):
    for row in range(2):
        ax = axs[row, col]
        pcm = ax.pcolormesh(np.random.random((20, 20)) * (col + 1),
                            cmap=cm[col])
    fig.colorbar(pcm, ax=axs[:, col], shrink=0.6)
```

```python
fig, axs = plt.subplots(3, 3, constrained_layout=True)
for ax in axs.flat:
    pcm = ax.pcolormesh(np.random.random((20, 20)), cmap='viridis')

fig.colorbar(pcm, ax=axs[0, :2], shrink=0.6, location='bottom')
fig.colorbar(pcm, ax=[axs[0, 2]], location='bottom')
fig.colorbar(pcm, ax=axs[1:, :], location='right', shrink=0.6)
fig.colorbar(pcm, ax=[axs[2, 1]], location='left')
```

### Iris Example

```python
from sklearn.datasets import load_iris

iris = load_iris()
features = iris.data.T
```

```python
features.shape
```

```python
iris.target
```

```python
plt.scatter(features[0], features[1], alpha = 0.2,\
           s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

ax1.scatter(features[0], features[1], alpha = 0.2,\
           s=100*features[3], c=iris.target, cmap='viridis')
ax1.set_xlabel(iris.feature_names[0])
ax1.set_ylabel(iris.feature_names[1])
```

### plot Versus scatter: A Note on Efficiency

- While it doesn’t matter as much for small
amounts of data, as datasets get larger than a few thousand points, plt.plot can be
noticeably more efficient than plt.scatter. The reason is that plt.scatter has the
capability to render a different size and/or color for each point, so the renderer must
do the extra work of constructing each point individually. In plt.plot, on the other
hand, the points are always essentially clones of each other, so the work of determin‐
ing the appearance of the points is done only once for the entire set of data. For large
datasets, the difference between these two can lead to vastly different performance,
and for this reason, plt.plot should be preferred over plt.scatter for large
datasets.
