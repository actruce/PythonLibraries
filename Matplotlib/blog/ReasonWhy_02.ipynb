```python
# We did that all we want draw. But this is a kind of stupid approach, isn't it?
# Let's do the same thing with object-oriented style
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(x, y)
```

```python
ax1.set_xlabel('X')
fig
```

```python
ax1.set_ylabel('Y')
ax1.set_title('Simple X and Y')
fig
```
